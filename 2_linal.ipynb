### Линейная  независимость

Векторы линейно зависимы, если один из них линейно выражается через остальные

То есть 
1. Хотяб один из них не равен нулю
2. Сумма векторов с такими кэфами равна нулю  

Линейно независимы, если сумма всех векторов будет равна нулю только если все они равны друг друг  
Пример:  
$x_{1}=(2,1,2)$  
$x_{2}=(4,2,4)$  
$x_{3}=(1,2,3)$  
Поступим так  
Замечаем что первый и второй вроде как зависимы, избавляемся от третьего умножением его на ноль, а первые два на кэфы  
$2x_{1} - 1x_{2}+0x_{3} = 0$  
получается ноль => они линейно зависимы

$x_{1}=(2,1,2)$  
$x_{2}=(5,2,4)$  
$x_{3}=(1,2,3)$  
тут уже какие кэфы не подставляй, они не будут равны нулю => Линейно независимые

Как же на глаз определить зависимы они или нет? Есть два стула:
1. Метод Гаусса - сложно и ниочем, нам не нужон

### Норма
Длина вектора обобщается с помощью Нормы  
Норма это фукнция $||x||$ от вектора которая должна удовлетворять 3 требованиям:  
* если $||x||$ = 0, то $x$ = 0   
* $||x + y|| \le  ||x|| + ||y||$ 
* $|| \alpha  * x || = |\alpha| * ||x|| $  

Векторное пространство с нормой - Нормированное

<h3><center>Евклидова норма</center></h3>
$$ ||x||_{2} = \sqrt{\sum_{i=1}^n x^2_{i}}$$

<h3><center>Манхэттенская норма</center></h3>
$$ ||x||_{1} = \sum_{i=1}^n |x_{i}|$$

Рассмотрим пример: Найти отношение Евклидовой нормы к Манхэттенской вектора (2,-1,-2)  
  
$$ \frac{\sqrt{2^2 + (-1)^2 + (-2)^2}}{|2|+ |-1| + |-2|} = \frac{3}{5} $$

#### Метрика \\ расстояние
Расстояние между векторами х и у  
$ p(x,y) = ||x-y|| $

<h3><center>Евклидова Метрика</center></h3>  
$$ p(x,y) = \sqrt{ \sum^n_{i=1}(x_{i}- y_{i})^2 } $$

<h3><center>Манхэттенская Метрика</center></h3>  
$$ p(x,y) = \sum^n_{i=1} |x_{i}- y_{i}|  $$

### Скалярное произведение
$\langle x,y\rangle = \sum^n_{i=1}x_{i}y_{i}$  
  
$ \langle x, y \rangle = ||x|| * ||y|| * \cos(x|y) $   
  
$| \bar{a} |* | \bar{b} |*\cos \alpha $
  
#### Норма
$||x|| = \sqrt{\langle x,x\rangle} $

#### Расстояние
$p(x,y) = ||x-y|| = \sqrt{\langle x-y,x-y\rangle}$
  
#### Угол  
$ \cos(x|y) = \frac{ \langle x, y \rangle } {||x|| * ||y|| }  $ 
  
для Параллельных векторов косинус = 1, для перпендикулярных = 0
