# Основы математического Анализа

## 1. Понятие функции действительной переменной

f(x) - функция  
x - аргумент

Числа бывают:
* $\mathbb N$ - Натуральные  {0,1,2,3,4...$+\infty$}       => [0,$+\infty$)
* $\mathbb Z$ - Целые        {$-\infty$...-1, 0, 1...$+\infty$} => ($-\infty$,$+\infty$)
* $\mathbb Q$ - Рациональные - те что можно представить дробью {...$\frac{1}{2}$,$\frac{1}{1}$,$\frac{2}{1}$...}
* Иррациональные числа - те что нельзя представить дробью {...$\sqrt{2}$,$e$,$\pi$,...}
* $\mathbb R$ - Действительные (вещественные) - все => {$-\infty$,$+\infty$}

$\in$ - принадлежит  
x$\in\mathbb R$ - x принадлежит R (x=любое)

{...} - множество, состоит из некоторого кол-ва чисел, например {1,3,5,7}

функция f(x) по сути это перевод одного множества в другое по заданной формуле  
аргумент х принадлежит одному множеству $x\in${1,2,3,4,5}  
задаем формулу  $y = f(x) = x^2$  
получаем новое множество $y\in${1,4,9,16,25}

```python
x = {1, 2, 3, 4, 5}
print('x = {1, 2, 3, 4, 5}')
y = {num**2 for num in x}
print('y =', y)
```

$\cup$ - объединение  
{0,1,2} $\cup$ {1,2,3} = {0,1,2,3} 

```python
a = {0,1,2}
b = {1,2,3}
print(a.union(b))
```

$\setminus$ - разность множеств (первое без всех вторых)  
{0,1,2} $\setminus$ {2} = {0,1}

```python
a = {0,1,2}
b = {2}
print(a.difference(b))
```

D(f) - Область определения функции (D - determine - определять) => какие значения может принимать Х  
E(f) - Область значений функции (E - expected? - ожидаемая) => какие значения может принимать функция f(x) == y  
  
примеры:  
$f(x) = \frac{1}{x-1}$  
D(f)= R \\ {1} (Все кроме 1, так как знаменатель не может быть равен нулю)  
E(f)= R \\ {0} (Дробь с единицей в числителе никогда не будет равна нулю)  
  
$f(x) = 2^x$  
D(f) = R (любая)  
E(f) = $(0,+\infty)$ (2 в любой степени не будет равна нулю, либо отрицательному числу)

График функции = подставить в формулу f(x) все значения X  
Обозначить по оси X значения X  
по оси Y все значения f(x)  
Пример:  
$f(x)=x^2+1$, $x\in${-10,10}

```python
x = np.arange(-10,10)
y = x**2 + 1

def graph(x,y):
    plt.plot(x, y)
    plt.xlabel('X')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()
    
graph(x,y)
```

```python
x = np.arange(-100,100)
y = x/(x+1)
graph(x,y)
```

### Непрерывность функции. Разрывы
Если у функции нет точек разрыва - она непрерывная.  
Разрыв - резкий скачок значений, выбивающийся из последовательности.  
#### Пример неразрывной функции: $f(x) = sin(x)$

```python
x = np.arange(-10, 10, 0.01)
y = np.sin(x)
graph(x,y)
```

#### Пример функции с разрывами
(matplotlib соединяет разрывы, но видно где они есть)  
$f(x) = arctan(\frac{1}{x})$  
  
$f(x) = \frac{|2x+5|}{2x+5}$

```python
x = np.arange(-10, 10, 0.01)
y = np.arctan(1/x)
graph(x,y)
```

```python
x = np.arange(-25, 25, 0.01)
y = (np.absolute(2*x+5)/(2*x+5))
graph(x,y)
```

#### Гладкость функции
Гладкие функции которые без углов, например f(x)=sin(x)  
Пример негладкой функции $f(x) = -|x|$

```python
x = np.arange(-100, 100, 0.01)
y = -np.absolute(x)
graph(x,y)
```
