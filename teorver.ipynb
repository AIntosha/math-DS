## Условная вероятность
$P(A+B)=P(A) + P(B) -P(AB)$  
$P(B\setminus A) = P(B) - P(AB)$  
У независимых $ P(AB) = P(A)P(B) $  
$ P(A|B) = \frac{P(AB)}{P(B)} $  
формула полной вероятности $ P(A) = P(A|B)P(B) + P(A| \bar B)P(\bar B) $  
формула Байеса$ P(A|B) = \frac{P(A)P(B|A)}{P(B)} $ 

## Дискретная случайная величина
Дискретная случайная величина (бросок кубика {1,2,3,4,5,6} = vec(p1,p2,p3,p4,p5,p6) , монетки) --  
Имеет счетное кол-во значений $A=\{ a_{1},a_{2},a_{3},a_{4},a_{5},a_{6}\}$ с вероятностями $p_{1},p_{2},p_{3},p_{4},p_{5},p_{6}$ и общей суммой вероятностей = 1  
тогда  
Если два исхода у монетки, получится:  
$P(X=1) = p$  
$ P(X=0) = 1-p$
  
Если хотим посчитать подряд (**сумма независимых бинарных случайных величин**):  
Вероятность попадания мяча в баскетбольное кольцо равна $p$, имеется $n$ попыток, а число попаданий равно $X$.  
$ P(X = n) = p^n $  
Если хотим попасть подряд 4 раза с вероятностью в 0.2 ::: $P(X=4) = 0.2^4 = 0.0016 = 0.16$%
  
В общем случае:  
$ P(X=k) = C^k_{n}p^k(1-p)^{n-k} $  
где:  
$ C^k_{n} = \frac{ n! }{k!(n-k)!} $  
итоговая универсальная формула:  
$ P(X=k) = \frac{ n! }{k!(n-k)!}p^k(1-p)^{n-k} $  
Величина X называется **Биномиальной**. Биномиальное распределение имеет два параметра n и $p \in [0,1]$

#### Счетчики
Примеры: Количество раз каждое слово встречается в книге,  
Количество автобусов проезжающих за час,  
Число Радиоактивных распадов, которые улавливает счетчик Гейгера.  

$X$ - Кол-во использования слова в тексте  
Вероятность того, что $X$ равно $k$ описывается **Распределением Пуассона**:  
$$ P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$$ 
$\lambda > 0$  математическое ожидание случайной величины (среднее количество событий за фиксированный промежуток времени)

## Непрерывные случайные величины
То есть те где у нас бесконечности  
Задаются двумя способами: 
1. Функция Распределения
2. Плотность Распределения
  
#### Функция распределния
Идет от нуля до единицы  
Всегда возрастает  
Функции разных величин похожи друг на друга, поэтому неудобны

```python
mu = 2.0
sigma = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=10)
x = np.linspace(0,4,100)
cdf = norm_rv.cdf(x)
plt.plot(x, cdf);
```

#### Плотность распределения
Считается через интеграл  
Разные функции отличаются друг от друга, поэтому удобны на практике

$\sigma$ - среднеквадратичное отклонение (примерный разброс +-)
  
**Нормальная случайная величина (Гауссовская)**
